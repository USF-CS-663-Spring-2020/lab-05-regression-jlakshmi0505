```python
#importing libraries
import pandas as pd
import seaborn as sns 
from sklearn.linear_model import LinearRegression
import matplotlib as plt
%matplotlib inline
```

```python
#Reading data
df = pd.read_csv("Credit.csv")
df.head()
```

#  Income vs Limit

```python
df.plot(kind='scatter', x='Income', y='Limit')
```

```python
plot = sns.lmplot(y='Limit', x='Income', data=df)
```

# Residual(Income) vs Limit

```python
#residual plot
plot1 = sns.residplot('Income', 'Limit', data=df)
plot1.set(xlabel='Residual', ylabel='Limit')
```

```python
reg = LinearRegression()
reg.fit(df[['Income']], df[['Limit']])
reg
```

```python
#slope
reg.coef_
```

```python
#intercept
reg.intercept_
```

```python
#Income is good predictor for Limit using Simple linear regression (y = 51.87*x + 2389)
#As variance is not changing ,so its a good predictor
```

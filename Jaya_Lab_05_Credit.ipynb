```python
#importing libraries
import pandas as pd
import seaborn as sns 
from sklearn.linear_model import LinearRegression
import matplotlib as plt
%matplotlib inline
```

```python
#Reading data
df = pd.read_csv("Credit.csv")
df.head()
```

#  Income vs Limit

```python
df.plot(kind='scatter', x='Limit', y='Income')
```

```python
plot = sns.lmplot(y='Income', x='Limit', data=df)
```

# Residual(Income) vs Limit

```python
#residual plot
plot1 = sns.residplot('Limit', 'Income', data=df)
plot1.set(xlabel='Limit', ylabel='Residual')
```

```python
reg = LinearRegression()
reg.fit(df[['Limit']], df[['Income']])
reg
```

```python
#slope
reg.coef_
```

```python
#intercept
reg.intercept_
```

```python
#Income is good predictor for Limit using Simple linear regression (y = 0.012x - 12.055)
```

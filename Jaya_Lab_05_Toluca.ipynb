```python
#importing libraries
import pandas as pd
import seaborn as sns 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('toluca.csv')
df.head()
```

```python
df.plot(kind='scatter', x='workHours', y='lotSize')
```

```python
plot = sns.lmplot(y='lotSize', x='workHours', data=df)
```

```python
#residual plot
plot1 = sns.residplot('workHours', 'lotSize', data=df)
plot1.set(xlabel='Residual', ylabel='Lot Size')

#random distribution around the regression line,so good
```

```python
reg = LinearRegression()
reg.fit(df.drop('lotSize', axis=1), df.iloc[:,0])
reg
```

```python
#slope
reg.coef_
```

```python
#intercept
reg.intercept_
```

```python
#WorkHours is good predictor for lotSize using Simple linear regression ( y = 0.23*x - 1.85 )
#As variance is not changing ,so its a good predictor
```

```python
#importing libraries
import pandas as pd
import seaborn as sns 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('toluca.csv')
df.head()
```

```python
df.plot(kind='scatter', x='lotSize', y='workHours')
```

```python
plot = sns.lmplot(y='workHours', x='lotSize', data=df)
```

```python
#residual plot
plot1 = sns.residplot('lotSize', 'workHours', data=df)
plot1.set(xlabel='Lot Size', ylabel='Residual')

#random distribution around the regression line,so good
```

```python
reg = LinearRegression()
reg.fit(df.drop('workHours', axis=1), df.iloc[:,1])
reg
```

```python
#slope
reg.coef_
```

```python
#intercept
reg.intercept_
```

```python
#WorkHours is good predictor for lotSize using Simple linear regression ( y = 3.57x + 62.36 )
```
